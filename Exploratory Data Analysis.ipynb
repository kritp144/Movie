 ### LIBRARIES

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### MERGING DATAFRAMES

```python
# Load the dataframes

df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
# Load 2000 dataframe

df_2000.head()
```

```python
# Load 2001 dataframe

df_2001.tail()
```

```python
# Combine both dataframes

df_combined = pd.concat([df_2000, df_2001])
display(df_combined.head(),df_combined.tail())
```

### DATA EXPLORATION

```python
# Check info on dataset

df_combined.info()
```

##### Filter Movies without Financial Info

```python
# Create a filter to find relative financial info

rev_filter= df_combined['revenue'] > 0
budg_filter= df_combined['budget'] > 0 
```

```python
# Apply the filters in a new dataframe

new_df = df_combined[revenue_filter | budget_filter]
n_movies = new_df.shape[0]

print(n_movies,
      'movies have relevant financial info regarding the budget or revenue')
```

##### How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Check Value Counts to Determine the number of movies in each 
# certification category

df_combined['certification'].value_counts()
```

##### What is the average revenue per certification category?

```python
# Group revenue by certification category

new_df.groupby(['certification'])['revenue'].mean()
```

##### What is the average budget per certification category?

```python
# Group revenue by certification category

new_df.groupby(['certification'])['budget'].mean()
```
