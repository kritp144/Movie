## Loading the Data

```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Load the files

basics_url = "https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
```

```python
# read the data in a dataframe

basics = pd.read_csv(basics_url,sep='\t', low_memory=False)
akas = pd.read_csv(akas_url,sep='\t', low_memory=False)
ratings = pd.read_csv(ratings_url,sep='\t', low_memory=False)
```

## Basics Data Cleaning

```python
basics.head()
```

```python
# Replace \N to np.nan

rep = {'\\N': np.nan}

basics = basics.replace(rep)
```

```python
basics.head()
```

```python
# Eliminating movies with a null value for runtimeMinutes and genre

basics.dropna(subset=['runtimeMinutes'], inplace=True)
basics.dropna(subset=['genres'], inplace=True)
```

```python
# Confirm removal of missing rows

basics.isna().sum()
```

```python
# Keep only titleType == Movie

movie_filter = basics['titleType']== 'movie'

basics = basics[movie_filter]

basics.head()
```

```python
# Keep startyear 2000 - 2022


yearmin_filter = basics['startYear'] >= '2000'
yearmax_filter = basics['startYear'] <= '2021'
basics = basics[yearmin_filter & yearmax_filter]

print('min start year', basics['startYear'].min())
print('max start year', basics['startYear'].max())
```

```python
# Remove Documentaries from Genres

is_doc = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_doc]
basics.head()
```

## AKAs Data Cleaning

```python
akas.head()
```

```python
# Keep only US entries

region_filter = akas['region']== 'US'

akas = akas[region_filter]

akas.head()
```

```python
# Replace \N to np.nan

akas = akas.replace(rep)
akas.head()
```

```python
# Filter basics to only include US region by akas

keepers =basics['tconst'].isin(akas['titleId'])

basics = basics[keepers]
basics.head()
```
