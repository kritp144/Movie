### Required Transformations

```python
# Show the title basics data
df.head()
```

```python
# Create a new column with a list of genres

df['genres_split'] = df['genres'].str.split(',')
df.head()
```

```python
# Separate the list of genres to a list of strings in each row

exploded_genres = df.explode('genres_split')
exploded_genres.head()
```

```python
# Get Unique genres

unique_genres = sorted(exploded_genres['genres_split'].unique())
```

```python
# Save tconst and genres_split in a new data frame

title_genres = exploded_genres[['tconst', 'genres_split']]
title_genres.head()
```

```python
# Making the genre mapper dictionary

genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
# Make a new intiger genre_id 

exploded_genres['genre_id'] = exploded_genres['genres_split'].map(genre_map)


# Drop unnecessary columns
exploded_genres = exploded_genres.drop(columns=['genres_split','originalTitle',
                                                'isAdult', 'titleType',
                                                'genres'])

exploded_genres.info()
exploded_genres.head()
```

```python
# Calculate max string lengths for object columns

key_len = exploded_genres['tconst'].fillna('').map(len).max()
title_len = exploded_genres['primaryTitle'].fillna('').map(len).max()


# Create a schema dictonary using Sqlalchemy datatype objects

df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}
```

### Genres Table

```python
# Create the new Genre table

genre_lookup = pd.DataFrame({'genre_id': genre_map.values(),
                             'genre_name': genre_map.keys()})

genre_lookup.head()
```

### Title Genres Table

```python
# Create the new Title Genre table

genres_title = exploded_genres[['tconst','genre_id']]
genres_title.head()
```

### Title Basics Table

```python
# Drop duplicates in tconst

exploded_genres = exploded_genres.drop_duplicates(subset= ['tconst'])
exploded_genres.head()
```

### TMDB table

```python
tmdb_df = tmdb_df[['imdb_id','revenue','budget','certification']]
tmdb_df.head()
```

### Save to SQL

```python
# Save title_basics to sql with dtype and index=False

exploded_genres.to_sql('title_basics', engine, dtype=df_schema,
                       if_exists='replace',index=False)
```

```python
# Save genres to sql with dtype and index=False
genre_lookup.to_sql('genres', engine, dtype=df_schema,
                    if_exists='replace',index=False)
```

```python
# Save title_ratings to sql with dtype and index=False
ratings_df.to_sql('title_ratings', engine, dtype=df_schema,
                    if_exists='replace',index=False)
```

```python
# Save tmdb_data to sql with dtype and index=False
tmdb_df.to_sql('tmdb_data', engine, dtype=df_schema,
                    if_exists='replace',index=False)
```

```python
# Save title_genres to sql with dtype and index=False
genres_title.to_sql('title_genres', engine, dtype=df_schema,
                    if_exists='replace',index=False)
```
